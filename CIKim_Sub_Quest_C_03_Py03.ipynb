문제. 사용자로부터 두 개의 정수와 연산자(+,-,*,/,**)를 입력받아 계산 결과를 출력하는 프로그램을 만들어주세요


 조건
사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다
정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력하며 다시 입력받기를 시도합니다.
나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다.
사용자가 지원하지 않는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다.
math 모듈을 사용하여 제곱 연산(**)을 지원합니다.
계산기를 통해 계속 계산을 할 것인지 입력받습니다.

```python
import math

def calculator():


    try:
    # 예외가 발생할 수 있는 코드 블록 시작
        # 입력 받기


        num1 = int(input("첫 번째 정수: "))
        # input()으로 문자열을 받아서 int()로 정수 변환 후 num1에 저장

        num2 = int(input("두 번째 정수: "))
        # input()으로 문자열을 받아서 int()로 정수 변환 후 num2에 저장

        operator = input("연산자 (+, -, *, /, **): ")
        # 연산자를 문자열로 입력받아 operator 변수에 저장

        # 0으로 나누기 체크


        if operator == '/' and num2 == 0:
        # 연산자가 '/'이고 두 번째 숫자가 0인지 확인
            raise ZeroDivisionError("0으로 나눌 수 없습니다")
            # ZeroDivisionError 예외를 발생
        operations = {
        # 딕셔너리를 만들어 연산자별 계산 결과를 저장
            '+': num1 + num2,
            '-': num1 - num2,
            '*': num1 * num2,
            '/': num1 / num2,

            '**': math.pow(num1, num2)
            # 키 '**': 값은 math.pow()로 num1의 num2 제곱
        }

        if operator not in operations:
        # operator가 딕셔너리의 키에 없으면
            raise ValueError(f"'{operator}'는 지원하지 않는 연산자입니다")
            # ValueError 예외를 발생

        result = operations[operator]
        # 딕셔너리에서 operator를 키로 사용해 계산 결과를 가져와 result에 저장

        print(f"결과: {num1} {operator} {num2} = {result}")


    except ValueError as e:
    # ValueError가 발생하면  (e에 예외 저장)
        print(f"입력 오류: {e}")
        # 에러 메시지를 출력

    except ZeroDivisionError as e:
    # ZeroDivisionError 처리
        print(f"연산 오류: {e}")
        # 에러 메시지를 출력

    except Exception as e:
    # 위에서 잡지 못한 예외
        print(f"예상치 못한 오류: {e}")
        # 에러 메시지를 출력
```

```python
calculator()
```

```python
def calc():
    result = None

    try:
        work = input("기호를 입력하세요(+, _, *, /, ^) :")
        m = int(input("정수를 입력하세요: "))
        n = int(input("정수를 입력하세요: "))

        if work == "+":
            result = m + n
        elif work == "-":
            result = m - n
        elif work == "*":
            result = m * n
        elif work == "/":
            result = m / n
        elif work == "^":
            result = m ** n
        else:
            print("잘못된 기호입니다..")

    except ValueError as e:
        print("잘못된 입력입니다. 정수를 입력해주세요.", e)
    except ZeroDivisionError as e:
        print("0으로 나눌 수 없습니다.", e)

    return result
```

```python
calc()
```
